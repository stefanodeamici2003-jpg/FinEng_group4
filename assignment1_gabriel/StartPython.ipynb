## Question 3

the function diverges so I take -1 in the exponential 

```python
from scipy.optimize import minimize_scalar

mu = 1

def f(x):
    return -(x * np.exp(-(x - mu)**2)/2)  

res = minimize_scalar(f)
x_max = res.x
y_max = -res.fun
x_max
```

## Question 4

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-2*np.pi, 2*np.pi, 400)
y = np.sin(x)

plt.figure()
plt.plot(x, y)
plt.title("y = sin(x)")
plt.show()
```

```python
rho_vals = [1/2, -1/np.sqrt(2)]
x = np.linspace(-2, 2, 400)

for rho in rho_vals:
    X, Y = np.meshgrid(x, x)
    Z = X**2 - 2*rho*X*Y + Y**2 - 1
    plt.figure()
    plt.contour(X, Y, Z, levels=[0])
    plt.title(f"rho = {rho}")
    plt.gca().set_aspect('equal')
    plt.show()
```
